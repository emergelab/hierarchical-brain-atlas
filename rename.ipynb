```python
from pathlib import Path
```

```python
directories = list(Path("./").glob("AA*"))
```

```python
for directory in directories:
    if directory.is_dir():
        splits = directory.name.split("+")
        if len(splits) > 1:
            new_directory_name = splits[0] + "Parcels" + splits[1].capitalize()
        else:
            new_directory_name = splits[0] + "Parcels"

        # Create new directory if it doesn't exist
        new_directory = directory.parent / new_directory_name
        new_directory.mkdir(exist_ok=True)

        print(f"Moving files from {directory.name} to {new_directory_name}")

        for file in directory.glob("*"):
            splits = file.name.split("_")
            splits[0] = f"atlas-{new_directory_name}"
            new_file_name = "_".join(splits)

            # Create the new file path
            new_file_path = new_directory / new_file_name

            # Move the file using rename()
            try:
                file.rename(new_file_path)
                print(f"Moved {file.name} to {new_file_path}")
            except Exception as e:
                print(f"Error moving {file.name}: {e}")
```
